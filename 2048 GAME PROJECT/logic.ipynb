```python
import random

### 4* 4 matrix with 0(empty pos) in it

def start_game():
    mat=[]
    for i in range(4):
        mat.append([0]*4)
    return mat 

### add random 2 at position which is empty or 0 

def add_new_2(mat):
    r=random.randint(0,3) ##random row
    c=random.randint(0,3)   ##radnom col
    while (mat[r][c]!=0): ### produce random row and col till we dont find an empty space
        r=random.randint(0,3)
        c=random.randint(0,3)  
    mat[r][c]=2
    
### current state of game , we need to check at every point what are possibilities
### like we won , lost , game still can be played

def get_current_state(mat): ### returns 3 things won ,lost ,still play
    for i in range(4):
        for j in range(4):
            if mat[i][j]==2048:
                return 'WON'
    ## if any o present
    for i in range(4):
        for j in range(4):
            if mat[i][j]==0:
                return 'GAME NOT OVER'
    ### evn if any adj cell has same no we are still in game
    ### every row and col except last row and column
    for i in range(3):
        for j in range(3):
            if (mat[i][j]==mat[i+1][j] or mat[i][j]==mat[i][j+1]):
                return 'GAME NOT OVER'
    ### last row
    for j in range(3):
        if mat[3][j]==mat[3][j+1]:
            return 'GAME NOT OVER'
    ### last col
    for i in range(3):
        if mat[i][3]==mat[i+1][3]:
            return 'GAME NOT OVER'
    ### lost
    return 'LOST'


### COMPRESS FUNC--- ALL NON ZERO NO MOVES TO THE MOVEMENT SIDE
def compress(mat):
    changed=False
    new_mat=[]
    for i in range(4):
        new_mat.append([0]*4)
    for i in range(4):
        pos=0
        for j in range(4):
            if mat[i][j]!=0:
                changed=True
                new_mat[i][pos]=mat[i][j]
                if j!=pos:
                    changed=True
                pos+=1
    return new_mat,changed
                
### merging two same consecutive/adj no
def merge(mat):
    changed=False
    for i in range(4):
        for j in range(3):
            if mat[i][j]==mat[i][j+1] and mat[i][j]!=0:
                changed=True
                mat[i][j]=mat[i][j]*2
                mat[i][j+1]=0 ###vvi
    return mat,changed
    
                    

### replaing a matrix
def reverse(mat):
    new_mat=[]
    for i in range(4):
        new_mat.append([])
        for j in range(4):
            new_mat[i].append(mat[i][4-j-1])
            
    return new_mat

###transpose
def transpose(mat):
    new_mat=[]
    for i in range(4):
        new_mat.append([])
        for j in range(4):
            new_mat[i].append(mat[j][i])
    return new_mat


### FOCUS ON LEFT ONLY ADN FOR OTHER MOVEMENTS WE WILL GENERALIZE
### left---- comp,merge,compree
### right--- rev ,comp,merge,compree,rev
### up-- tanspose, comp,merge,compree,transpose
### down -- tanspose,rev, comp,merge,compree,rev,transpose


### IF ANY CHANGE OR MOVEMENT OR COMPRESSION OR MERGING HAPPENED THEN ONLY NEW RANDOM IS ADDED
### SO WITH RETURNING MAT WE ALSO NEED TO RETURN IF SOME CHANGE HAPPEND OR NOT
 
def move_up(grid):
    transpose_grid=transpose(grid)
    new_grid,changed1=compress(transpose_grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid,temp=compress(new_grid)
    final_grid=transpose(new_grid)
    return final_grid,changed
    
def move_down(grid):
    transpose_grid=transpose(grid)
    reverse_grid=reverse(transpose_grid)
    new_grid,changed1=compress(reverse_grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid,temp=compress(new_grid)
    new_grid=reverse(new_grid)
    final_grid=transpose(new_grid)
    return final_grid,changed

def move_left(grid):
    new_grid,changed1=compress(grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid,temp=compress(new_grid)
    return new_grid,changed

def move_right(grid):
    reverse_grid=reverse(grid)
    new_grid,changed1=compress(reverse_grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid,temp=compress(new_grid)
    final_grid=reverse(new_grid)
    return final_grid,changed

```
